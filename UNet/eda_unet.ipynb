```python
from PIL import Image
import os
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Visualization Comparison between input_img and mask_true and mask_pred -> phantom_20
input_img = Image.open('data/imgs/phantom_20/12.jpg')
mask_true = Image.open('data/masks/phantom_20/12.jpg')
mask_pred = Image.open('output/phantom_20/12.jpg')

imgs = [input_img, mask_true, mask_pred]
titles = [f'input_img: {input_img.size}', f'mask_true: {mask_true.size}', f'mask_pred: {mask_pred.size}']
plt.figure(figsize=(20,10))
for i, (img, title) in enumerate(zip(imgs, titles)):
    plt.subplot(1, 3, i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(title)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(mask_true)
plt.imshow(mask_pred, alpha=0.3)
plt.axis('off')
plt.show()
```
